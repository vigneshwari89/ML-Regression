```python
import xgboost as xg # xgboost is a library .
```

```python
dataset=pd.read_csv("Assignment insurance_pre.csv") ##dataset is called used to read and assigned the dataset variable
```

```python
dataset # run the dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True) #state is calssification. It means nominal-one hot encoding convented to numaric data then assigned the dataset
#variable
```

```python
dataset # run the dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age','bmi','children', 'sex_male','smoker_yes']] #independent is split the input
```

```python
independent
```

```python
dependant=dataset[['charges']] #dependent is split the output
```

```python
dependant #run the output
```

```python
from sklearn.model_selection import train_test_split #It is model selection and split train and test
x_train,x_test,y_train,y_test=train_test_split(independent,dependant,test_size=0.30,random_state=0)  # it is called particular function to 
# assigned perameter input and ouput passing to the variable
```

```python
from xgboost import XGBRegressor # it is two line procedure.xg boost is a library it used XGBRegressor algorthim.
regressor = XGBRegressor(n_estimators=10) #regressor is a varibale to assign the parameters
regressor.fit(x_train, y_train) #regressor.fit is function call to assign traing set input and output
```

```python
y_pred=regressor.predict(x_test)  # The check evaluation metrics. y_pred is variable, regressor is model creation, predict is 
# application function name and (x-test) is input data passing in evalation metrics.
```

```python
from sklearn.metrics import r2_score   # R2 value check
r_score=r2_score(y_test,y_pred) # It is r_score varible name and r2_score to assigned Actual data and pred data check
```
